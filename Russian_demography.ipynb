```python
import seaborn as sns
sns.jointplot(x='urbanization' , y='npg' , data=df , size=15)
```

Вывод: чтобы Русский Народ не вымер, его нужно поселить в среду с урбанизацией в 30-40%

```python
#отсортируем по странам и годам
df1 = df.sort_values(by=['region', 'year'])
df1
```

```python
#индексы сбились, нужно подправить
df1.index = np.arange(len(df1))
df1.head(30)
```

```python
#Посмотрим, что происходит в 2017, самом последнем году датафрейма.
df17 = df1.loc[df1['year'] == 2017]
df17 = df17.drop(['birth_rate', 'death_rate', 'gdw', 'year'], axis=1)
df17.rename(columns={'npg': 'npg17', 'urbanization': 'urbanization17'}, inplace=True)
df17
```

```python
df90 = df1.loc[df1['year'] == 1990]
df90 = df90.drop(['birth_rate', 'death_rate', 'gdw', 'year'], axis=1)
df90.rename(columns={'npg': 'npg90', 'urbanization': 'urbanization90'}, inplace=True)
df90
```

```python
#объединяем
df28 = df17.merge(df90, how='outer')
df28
```

```python
#смотрим изменения за 28 лет
df28["npg_change"] = df28["npg17"] - df28["npg90"]
df28["urbanization_change"] = df28["urbanization17"] - df28["urbanization90"]
df28 = df28.drop(['npg17', 'urbanization17', 'npg90', 'urbanization90'], axis=1)
df28
```

```python
#какая связь между изменением урбанизации и зменением прироста
sns.jointplot(x='urbanization_change' , y='npg_change' , data=df28 , size=15)
```

Ничего не понятно

```python
df28["cor"] = df28["npg_change"] / df28["urbanization_change"]
df28.cor.median()
```

Так-то лучше. Чтобы сохранить Русский Народ, нужно разрушать города. Чуть-чуть.
